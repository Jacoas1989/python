Puedes identificar fácilmente estos valores si intentas dibujar el cuadrado a partir de su diagonal. Si andas perdido, prueba de dibujarlo en un papel, ¡seguro que lo verás mucho más claro! Además recuerda que puedes utilizar la función abs() para saber el valor absolute de un número



```python
import math
math.sqrt(9)

class Punto:
    x=0
    y=0
    
    def __init__(self,x=0,y=0):
        self.x=x
        self.y=y
        print(x,y)
        
    def __str__ (self):
        return f"El punto que aparece por pantalla es el ({self.x} ,{self.y})"
    
    def cuadrante(self):
        if self.x==0 and self.y!=0:
            print("sobre el eje Y")
        if self.x!=0 and self.y==0:
            print("sobre el eje x")
        if self.x==0 and self.y==0:
            print("Está sobre el origen")
        if self.x>0 and self.y>0:
            print("Cuadrante 1")
        if self.x>0 and self.y<0:
            print("Cuadrante 4")
        if self.x<0 and self.y>0:
            print("Cuadrante 2")
        if self.x<0 and self.y<0:
            print("Cuadrante 3")
            
    def vector(self,other):
        print (other.x-self.x, other.y-self.y)

    def distancia(self,other):
        print( math.sqrt(pow(other.x-self.x,2) + pow(other.y-self.y,2)) )
              
            
print("\nCreamos los siguientes puntos por pantalla:")            
punto_A = Punto(2,3)
punto_B = Punto(5,5)
punto_C = Punto(-3,-1)
punto_D = Punto(0,0)

print("\nImprimimos los puntos creados por pantalla:")
print(str(punto_A))
print(str(punto_B))
print(str(punto_C))
print(str(punto_D))

print("\nLos cuadrantes a los que pertenecen el punto A, C y D son:")
punto_A.cuadrante()
punto_C.cuadrante()
punto_D.cuadrante()

print("\nConsulta los vectores AB y BA:")
vector_AB = punto_A.vector(punto_B)
vector_BA = punto_B.vector(punto_A)

print("\nConsulta la distancia entre los puntos 'A y B' y 'B y A':")
distancia_AB = punto_A.distancia(punto_B)
distancia_BA = punto_B.distancia(punto_A)
```
